```python
products = [
    {"name": "Laptop", "category": "Electronics", "price": "1000"},
    {"name": "Mouse", "category": "Electronics", "price": 25},
    {"name": "Chair", "category": "Furniture", "price": "150"},
    {"name": "Desk", "category": "Furniture", "price": None},
    {"name": "Laptop", "category": "Electronics", "price": "1000"},
]
```

```python
Remove_products = []

for product in clean_products:
    if product["category"] != None and product["price"] != None:
        Remove_products.append(product)
```

```python
for product in Remove_products:
    product["price"] = float(product["price"])
```

```python
grouped_products = {}

for product in Remove_products:
    category = product["category"]
    
    if category not in grouped_products:
        grouped_products[category] = []
    
    grouped_products[category].append(product)
```

```python
for category in grouped_products:
    print("Category:", category)
    
    total = 0
    
    for item in grouped_products[category]:
        print("Product:", item["name"], " Price:", item["price"])
        total += item["price"]
    
    print("Total Price:", total)
    print()
```

```python
input_folder = "student_records"

for filename in os.listdir(input_folder):

    if filename.endswith(".csv"):

        print(f"\nProcessing {filename}...")

        filepath = os.path.join(input_folder, filename)

        try:
            with open(filepath, 'r') as file:
                reader = csv.DictReader(file)

                for row in reader:
                    try:
                        name = row["Name"]
                        marks = int(row["Marks"])

                        if marks < 0 or marks > 100:
                            raise InvalidDataError("Marks out of valid range (0-100)")

                        status = "Fail" if marks < 40 else "Pass"

                        with open("clean_results.csv", 'a', newline='') as out:
                            writer = csv.writer(out)
                            writer.writerow([name, marks, status])

                    except Exception as e:
                        print(f"Skipping row due to error: {e}")

        except FileNotFoundError:
            print(f"{filename} not found.")
```

Task 3:Project Directory Auditor

```python
%%writefile scanner_module.py
import os
from datetime import datetime, timedelta

def scan_directory(folder):
    data=[]
    now=datetime.now()
    limit=now-timedelta(days=30)

    for f in os.listdir(folder):
        p=os.path.join(folder,f)
        if os.path.isfile(p):
            m=datetime.fromtimestamp(os.path.getmtime(p))
            extra=0
            fine=0
            if m<limit:
                extra=(now-m).days-30
                fine=extra*10
            data.append([f,os.path.getsize(p),os.path.splitext(f)[1],m,extra,fine])
    return data

def generate_reports(d):
    t=datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    
    with open("audit_"+t+".txt","w") as a:
        for i in d:
            a.write(str(i)+"\n")

    with open("penalty_"+t+".txt","w") as p:
        for i in d:
            if i[5]>0:
                p.write(i[0]+" Fine Rs."+str(i[5])+"\n")
```

```python
ranked = sorted(
    employees,
    key=lambda x: (-x["score"], -x["experience"], x["name"])
)

ranked
```

```python
ranked_list = [
    {"rank": i+1, **emp}
    for i, emp in enumerate(ranked)
]

ranked_list
```

```python
name_sorted = sorted(ranked_list, key=lambda x: x["name"])
name_sorted
```

```python
def binary_search(data, target):
    low = 0
    high = len(data) - 1
    
    while low <= high:
        mid = (low + high) // 2
        
        if data[mid]["name"] == target:
            return data[mid]
        elif data[mid]["name"] < target:
            low = mid + 1
        else:
            high = mid - 1
            
    return "Employee Not Found"
```

Task 6:Order Discount Pipeline

```python
from functools import reduce
```

```python
orders = [
    {"customer": "Ali", "category": "Electronics", "price": 50000, "quantity": 1},
    {"customer": "Ahmed", "category": "Groceries", "price": 800, "quantity": 10},
    {"customer": "Zara", "category": "Electronics", "price": 20000, "quantity": 2},
    {"customer": "Usman", "category": "Clothing", "price": 2500, "quantity": 2}
]
```

```python
discount_rules = {
    "Electronics": 0.10,
    "Clothing": 0.20,
    "Groceries": 0.05
}
```

```python
discounted_orders = list(map(
    lambda o: {
        **o,
        "total_price": o["price"] * o["quantity"],
        "discounted_price": (
            o["price"] * o["quantity"]
            - (o["price"] * o["quantity"] * discount_rules[o["category"]])
        )
    },
    orders
))

discounted_orders
```

```python
filtered_orders = list(filter(
    lambda o: o["discounted_price"] >= 5000,
    discounted_orders
))

filtered_orders
```

```python
sorted_orders = sorted(
    filtered_orders,
    key=lambda o: o["discounted_price"],
    reverse=True
)

sorted_orders
```

```python
total_revenue = reduce(
    lambda acc, o: acc + o["discounted_price"],
    sorted_orders,
    0
)

total_revenue
```

```python
summary = {
    "Processed Orders": sorted_orders,
    "Total Revenue": total_revenue
}

summary
```

Task 7: Server Log Error Streamer

```python
import random
from datetime import datetime, timedelta

# Create a simulated large log file
log_file = "server.log"

levels = ["INFO", "WARNING", "ERROR", "CRITICAL"]
error_types = ["DatabaseError", "TimeoutError", "ConnectionError", "AuthError"]

start_time = datetime.now()

with open(log_file, "w") as f:
    for i in range(10000):  # simulate large file
        timestamp = start_time + timedelta(seconds=i)
        level = random.choice(levels)

        if level in ["ERROR", "CRITICAL"]:
            error = random.choice(error_types)
            f.write(f"{timestamp} - {level} - {error} - Something went wrong\n")
        else:
            f.write(f"{timestamp} - {level} - Routine log message\n")

print("Sample log file created.")
```

```python
class LogFileIterator:
    def __init__(self, filepath):
        self.file = open(filepath, "r")

    def __iter__(self):
        return self

    def __next__(self):
        line = self.file.readline()
        if not line:
            self.file.close()
            raise StopIteration
        return line.strip()
```

```python
log_iter = LogFileIterator("server.log")

for i, line in enumerate(log_iter):
    if i == 5:
        break
    print(line)
```

```python
def critical_error_stream(log_iterator):
    for line in log_iterator:
        if "ERROR" in line or "CRITICAL" in line:
            yield line
```

```python
log_iter = LogFileIterator("server.log")

for i, entry in enumerate(critical_error_stream(log_iter)):
    if i == 5:
        break
    print(entry)
```

```python
def extract_error_type(log_line):
    parts = log_line.split(" - ")
    if len(parts) >= 3:
        return parts[2]
    return None
```

```python
log_iter = LogFileIterator("server.log")
errors = critical_error_stream(log_iter)

error_types = (extract_error_type(line) for line in errors)

for i, e in enumerate(error_types):
    if i == 5:
        break
    print(e)
```

```python
from collections import defaultdict

def stream_error_counter(log_path):
    counts = defaultdict(int)

    log_iter = LogFileIterator(log_path)
    errors = critical_error_stream(log_iter)

    for line in errors:
        error_type = extract_error_type(line)
        if error_type:
            counts[error_type] += 1
            yield dict(counts)  # yield live snapshot
```

```python
counter_stream = stream_error_counter("server.log")

for i, snapshot in enumerate(counter_stream):
    if i % 1000 == 0:   # show every 1000 updates
        print(snapshot)
```

```python
from collections import defaultdict

def stream_server_errors(filepath):
    counts = defaultdict(int)

    with open(filepath, "r") as f:
        for line in f:
            if "ERROR" in line or "CRITICAL" in line:
                parts = line.strip().split(" - ")
                if len(parts) >= 3:
                    error_type = parts[2]
                    counts[error_type] += 1
                    yield error_type, counts[error_type]
```

```python
for error, count in stream_server_errors("server.log"):
    print(f"{error}: {count}")
```

Task 8:API Middleware Simulator

```python
import logging

# Configure logging
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[
        logging.FileHandler("app.log"),      # log to file
        logging.StreamHandler()              # log to console
    ]
)

# Create named logger
logger = logging.getLogger("api_logger")
```

```python
import time
from functools import wraps

def timing_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        
        result = func(*args, **kwargs)
        
        end_time = time.time()
        duration = end_time - start_time
        
        logger.info(f"{func.__name__} executed in {duration:.4f} seconds")
        return result
    
    return wrapper
```

```python
def log_io(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        logger.debug(f"Calling {func.__name__}")
        logger.debug(f"Inputs: args={args}, kwargs={kwargs}")
        
        result = func(*args, **kwargs)
        
        logger.debug(f"Output: {result}")
        return result
    
    return wrapper
```

```python
def require_role(required_role):
    def decorator(func):
        @wraps(func)
        def wrapper(user_role, *args, **kwargs):
            if user_role != required_role:
                logger.warning(
                    f"Access denied for role '{user_role}' on {func.__name__}"
                )
                return "Access Denied"
            
            logger.info(f"Access granted for role '{user_role}'")
            return func(user_role, *args, **kwargs)
        
        return wrapper
    return decorator
```

```python
@require_role("admin")
@log_io
@timing_decorator
def get_user_data(user_role, user_id):
    time.sleep(1)  # simulate processing
    return {"user_id": user_id, "name": "Ali raza"}
```

```python
@require_role("admin")
@log_io
@timing_decorator
def update_user(user_role, user_id, new_name):
    time.sleep(0.5)  # simulate processing
    return {"user_id": user_id, "updated_name": new_name}
```

task 9: Banking System

```python
class BankAccount:
    
    bank_name = "Python Bank"   # class attribute
    
    def __init__(self, owner, balance):
        self.owner = owner             
        self._account_type = "Savings"  
        self.__balance = 0             
        
        self.deposit(balance)           
    
    @property
    def balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}")
        else:
            print("Deposit must be positive")
    
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal must be positive")
        elif amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount
            print(f"Withdrew {amount}")
    
    @classmethod
    def change_bank_name(cls, new_name):
        cls.bank_name = new_name
    
    @staticmethod
    def bank_policy():
        return "Minimum balance must be positive."
    
    def __str__(self):
        return f"Owner: {self.owner}, Balance: {self.__balance}"
    
    def __repr__(self):
        return f"BankAccount('{self.owner}', {self.__balance})"
```
