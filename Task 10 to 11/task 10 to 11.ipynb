Task 10: Library Management System

```python
class Person:
    def __init__(self, name):
        self.name = name
    
    def get_limit(self):
        return 0
    
    def __str__(self):
        return f"Member: {self.name}"
```

```python
class Member(Person):
    def __init__(self, name):
        super().__init__(name)  
        self.borrowed_books = []
    
    def get_limit(self):
        return 2 
    
    def borrow(self, book):
        if len(self.borrowed_books) >= self.get_limit():
            print("Borrowing limit reached!")
        else:
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed {book.title}")
    
    def __len__(self):
        return len(self.borrowed_books)
```

```python
class PremiumMember(Member):
    def get_limit(self):   
        return 5
    
    def calculate_fine(self, days_overdue):
        return days_overdue * 1
```

```python
class RegularMember(Member):
    def calculate_fine(self, days_overdue):
        return days_overdue * 5
```

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        return f"{self.title} by {self.author}"
    
    def __lt__(self, other):
        return self.pages < other.pages
    
    def __eq__(self, other):
        return self.pages == other.pages
    
    def __add__(self, other):
        return self.pages + other.pages
```

```python
class FinePolicy:
    def calculate_fine(self, days):
        return days * 3

class NotificationMixin:
    def notify(self, message):
        print(f"Notification: {message}")

class SpecialMember(Member, FinePolicy, NotificationMixin):
    pass
```

```python
print(SpecialMember.mro())
```

```python
class Library:
    def __init__(self):
        self.books = []
    
    def add_book(self, book):
        self.books.append(book)
    
    def __len__(self):
        return len(self.books)
    
    def __getitem__(self, index):
        return self.books[index]
    
    def __setitem__(self, index, value):
        self.books[index] = value
    
    def __iter__(self):
        self._index = 0
        return self
    
    def __next__(self):
        if self._index < len(self.books):
            book = self.books[self._index]
            self._index += 1
            return book
        raise StopIteration
```

```python

book1 = Book("Computer", "Ali Raza", 300)
book2 = Book("Database", "Zohaib Ali", 500)


print("Is book1 smaller than book2?", book1 < book2)
print("Total pages of both books:", book1 + book2)

library = Library()
library.add_book(book1)
library.add_book(book2)

print("Total books in library:", len(library))

for book in library:
    print(book)

regular = RegularMember("Mahib")
premium = PremiumMember("Saif")

regular.borrow(book1)
premium.borrow(book2)

print("Regular fine:", regular.calculate_fine(3))
print("Premium fine:", premium.calculate_fine(3))

special = SpecialMember("Moaz")
print("Special fine:", special.calculate_fine(3))
special.notify("Your book is overdue!")
```

Task 11: Async Data Sync Dashboard

```python
import os
from dotenv import load_dotenv

load_dotenv()

class Config:
    API_KEY = os.getenv("API_KEY")
    SOURCE_DELAY = int(os.getenv("SOURCE_DELAY", 1))

print("Loaded API KEY:", Config.API_KEY)
```

```python
import asyncio

async def fetch_source_1():
    await asyncio.sleep(Config.SOURCE_DELAY)
    return {"source": "A", "value": 10}

async def fetch_source_2():
    await asyncio.sleep(Config.SOURCE_DELAY)
    return {"source": "B", "value": 20}

async def fetch_source_3():
    await asyncio.sleep(Config.SOURCE_DELAY)
    return {"source": "C", "value": 30}

async def fetch_all_sources():
    print("Fetching from all sources asynchronously...")
    results = await asyncio.gather(
        fetch_source_1(),
        fetch_source_2(),
        fetch_source_3()
    )
    return results
```

```python
from concurrent.futures import ThreadPoolExecutor

def heavy_calculation(data):
    return data["value"] * 2

def process_data(data_list):
    print("Processing data using parallel execution...")
    
    with ThreadPoolExecutor() as executor:
        results = list(executor.map(heavy_calculation, data_list))
    
    return results
```

```python
import threading

def log_results(results):
    print("Logging results in background thread...")
    print("Logged Data:", results)

def start_logger(results):
    thread = threading.Thread(target=log_results, args=(results,))
    thread.start()
    thread.join()
```

```python
class Dashboard:
    def __init__(self):
        self.api_key = Config.API_KEY

    async def run(self):
        raw_data = await fetch_all_sources()
        processed_data = process_data(raw_data)

        start_logger(processed_data)

        self.display_summary(processed_data)

    def display_summary(self, results):
        print("\n DASHBOARD SUMMARY")
        print("Total:", sum(results))
        print("Average:", sum(results)/len(results))
```

```python
dashboard = Dashboard()
await dashboard.run()
```
